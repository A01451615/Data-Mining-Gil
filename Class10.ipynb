```python
tra = []
for i in range(0, df.shape[0]):
    tra.append([ str(df.values[i,j])
                for j in range(0, df.shape[1])
               ]
              )
tra
```

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_aray = te.fit(tra).transform(tra)

encoded_df = pd.DataFrame(te_aray,  columns = te.columns_)
```

```python
encoded_df
```

```python
encoded_df.drop(['nan'],axis = 1, inplace = True)
encoded_df.columns
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
f_itemsets = fpgrowth(encoded_df,
                      min_support = 0.6,
                      use_colnames = True
                     )
f_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
a_rules = association_rules(f_itemsets,
                            metric = 'lift',
                            min_threshold = 1.5
                           )
a_rules
```
